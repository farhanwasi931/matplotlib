```python
#line plot
#time component is involved on x axis

x = [1,2,3,4,5]
y = [5,1,7,8,2]
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.title("line plot to understand syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x,y, color = "red")
plt.title("line plot to understand syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x,y, color = "red",marker = "o")
plt.title("line plot to understand syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x,y, color = "red",marker = "o",linestyle ='--')
plt.title("line plot to understand syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x,y, color = "red",marker = "o",linestyle ='--',linewidth = 3, markersize = 10)
plt.title("line plot to understand syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.grid()
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [5,1,7,8,2]
y2 = [7,8,3,5,1]

plt.plot(x,y1, color = "red",marker = "o",linestyle ='--',linewidth = 3, markersize = 6,label = 'line y1')
plt.plot(x,y2, color = "b",marker = "o",linestyle ='--',linewidth = 3, markersize = 10, label = 'line y2')

plt.xlabel("x data axis")
plt.ylabel("y data axis")
plt.title("multiple line in one plot")
plt.legend()
plt.show()
```

```python
#stacked histogram

data1 = np.random.normal(0,1,5000)
data2 = np.random.normal(0,1,5000)
```

```python
#pie chart
langs = ['c','python','java']
students = [20,100,40]

plt.pie(students,labels = langs, autopct = '%1.1f%%')
plt.title("pie chart")
plt.show()
```

```python
explode = (0.0,0.1,0.1)

langs = ['c','python','java']

students = [20,100,40]

plt.pie(students,labels = langs, autopct = '%1.1f%%',explode = explode ,shadow = True)
plt.title("pie chart")
plt.show()
```
